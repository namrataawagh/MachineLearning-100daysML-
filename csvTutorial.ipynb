# 3. Opening  a csv file from an URL

not important you can just refer to this if and when working w an URL

```python
import requests  # To handle HTTP requests
from io import StringIO # StringIO, used to treat strings as file-like objects.

url = "https://raw.githubusercontent.com/CourseMaterial/DataWrangling/main/flowerdataset.csv"
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
req = requests.get(url, headers=headers) #Sends an HTTP GET request to the specified url
data = StringIO(req.text) #StringIO allows pandas.read_csv() to treat this text as if it were an actual file

pd.read_csv(data)
```

# 4. Sep parameter
Sep() function defines the character that separates columns in your CSV file.

By default, pandas assumes that the separator is a comma (,), but sometimes CSV files use other delimiters like tabs (\t), semicolons (;), or pipes (|).
In that case you have to mention so

```python
# tsv file
pd.read_csv('titanic.tsv', sep='\t')
```

# 5 . Index_col parameter
if you don't want a certain column for analysis , you can convert it into an index column

**We initially faced here a UnicodeDecodeError; **occurs when pandas tries to read a CSV file that is not encoded in UTF-8 but in another encoding format (e.g., ISO-8859-1, Latin-1, or Windows-1252). This happens because pandas expects text to be in UTF-8 by default.

**Encoding** is the process of converting text (characters, symbols, emojis) into a format that computers can store and process as binary (0s and 1s).


```python
pd.read_csv('buisness.csv',encoding="ISO-8859-1" )
```

```python
# converting description column to index
pd.read_csv('buisness.csv',encoding="ISO-8859-1",index_col= 'description' )
```

# 7. use_cols parameter
Instead of loading the entire dataset, you can choose only the required columns, which improves efficiency.

```python
pd.read_csv('modified_placement_data.csv', usecols = ['cgpa','placement'])
```

# 8. Skip rows

```python
pd.read_csv('modified_placement_data.csv',skiprows=[2,5] ) #skips 2nd and 5th row
```

```python
pd.read_csv('modified_placement_data.csv', nrows = 10) #sets a limit of 10 rows
```

# 10. Converters
lets you apply custom functions to specific columns while reading a CSV file.

```python
def rename(ans):
  if ans == "Yes":
    return 'Y'
  else:
    return ans

```

```python
pd.read_csv('modified_placement_data.csv',converters ={'placement':rename})
```

# 11. na_value parameters
It tells pandas which values should be treated as NaN (Not a Number / missing values).

i.e If your dataset uses custom placeholders like "N/A", "missing", or "NULL" for missing values, you can convert them to NaN:

```python
pd.read_csv('buisness.csv',encoding="ISO-8859-1")
```

```python
pd.read_csv('buisness.csv',encoding="ISO-8859-1",na_values=['total'] ) # 'Total' was replaced by NaN
```
